```python
x=df1.drop('btc_market_price',axis=1)
y=df.btc_market_price
```

```python
print(x.isnull().sum())
```

```python
print(x['btc_total_bitcoins'].isnull().value_counts())
print(x['btc_trade_volume'].isnull().value_counts())
print(x['btc_blocks_size'].isnull().value_counts())
print(x['btc_median_confirmation_time'].isnull().value_counts())
print(x['btc_difficulty'].isnull().value_counts())
print(x['btc_transaction_fees'].isnull().value_counts())
```

```python
x['btc_total_bitcoins'].fillna(np.mean(x['btc_total_bitcoins']),inplace=True)
x['btc_trade_volume'].fillna(np.mean(x['btc_trade_volume']),inplace=True)
x['btc_blocks_size'].fillna(np.mean(x['btc_blocks_size']),inplace=True)
x['btc_median_confirmation_time'].fillna(np.mean(x['btc_median_confirmation_time']),inplace=True)
x['btc_difficulty'].fillna(np.mean(x['btc_difficulty']),inplace=True)
x['btc_transaction_fees'].fillna(np.mean(x['btc_transaction_fees']),inplace=True)
x.describe()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=1)
train_x.describe()
```

```python
lreg=LinearRegression()
lreg.fit(train_x,train_y)
pred=lreg.predict(test_x)
np.mean((pred-test_y)**2)
```
